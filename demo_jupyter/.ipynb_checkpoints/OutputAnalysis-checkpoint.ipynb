```python
import numpy as np
import scipy.misc
import matplotlib.pyplot as plt
import skimage.color
import os
%matplotlib inline
```

```python
def rgb2lab(image):
    yuv_color = skimage.color.rgb2lab(image)
    return yuv_color
image_path = 'D:\\PythonScript\\Comp5422\\github\\GANforColorQuantization\\tmp_output\\'
```

```python
image_generator = iter(os.listdir(image_path))
img_list = []
for image in image_generator:
    reading_path = image_path + image
    img = scipy.misc.imread(reading_path)
    img_list.append(img)
```

```python
plt.imshow(img_list[0])
```

```python
img_lab = rgb2lab(img_list[0])
img_lab[:5,:5,2] # 
```
